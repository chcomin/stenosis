```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv('stress_test_log.csv')

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(df['Elapsed_Seconds'], df['CPU_Temp_C'], label='CPU Temp (°C)', color='red')
plt.plot(df['Elapsed_Seconds'], df['GPU_Temp_C'], label='GPU Temp (°C)', color='green')

plt.title('Component Temperatures During Stress Test')
plt.xlabel('Time (Seconds)')
plt.ylabel('Temperature (°C)')
plt.legend()
plt.grid(True)
#plt.savefig('temperature_graph.png')
plt.show()
```
